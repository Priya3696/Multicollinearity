```python
#importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 1000) # shows all the columns
pd.set_option('display.float_format', lambda x: '%.4f' % x) # displays outputs with 1 decimal
pd.set_option('display.max_rows', 10000) # shows all the rows
%matplotlib inline
```

```python
df_adv = pd.read_csv("advertising.csv")
```

```python
df_adv.head()
```

```python
df_adv.columns
```

```python
import statsmodels.api as sm
X = df_adv[['TV','Radio','Newspaper']]
y = df_adv['Sales']
df_adv.head()
```

```python
X = sm.add_constant(X)
```

```python
X
```

```python
## fit a OLS model with intercept on TV and Radio

model= sm.OLS(y, X).fit()
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt
X.iloc[:,1:].corr()
```
